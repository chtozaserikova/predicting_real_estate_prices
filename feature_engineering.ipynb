```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('cian.csv')
df.isnull().sum(axis=0).sort_values()
```

перед удалением жилой площади найти квартиры, где nan значение в "Общая площадь" и прибавить (если есть) среднюю разницы общей и жилой площади к значению имеющейся жилой площади

```python
difference = round((df['Общая площадь']-df['Жилая площадь']).mean(), 1)
df.loc[(df['Жилая площадь'] != 0) & (df['Общая площадь'] == 0), 'Общая площадь'] = df['Жилая площадь']+difference
```

```python
col_to_drop = ['Срок сдачи', 'Жилая площадь', 'Unnamed: 0']
df.drop(col_to_drop, inplace=True, axis=1)
df
```

```python
df['Этаж'] = df['Этаж'].fillna(0)
df['Этажность дома'] = df['Этажность дома'].fillna(0)
df.loc[df['Этаж'] == 0]
```

```python
df.loc[df['Этажность дома'] == 0]
```

```python
values = {'Время до метро': round(df['Время до метро'].median())}
df.fillna(value = values, inplace = True)

to_drop = ['Метро', 'Ремонт', 'Этаж', 'Этажность дома', 'Цена', 'Общая площадь']
for dr in to_drop:
    df.drop(df[df[dr].isna()].index, inplace = True)
```

```python
df['Цена за метр'] = round(df['Цена'] / df['Общая площадь'])
```

```python
df.isnull().sum(axis=0).sort_values()
```

Теперь решаем вопрос с ванной комнатой

```python
df['Ванная'].value_counts()
```

```python
df['Ванная'].loc[(df['Ванная'] == '1 совмещенный') | (df['Ванная'] == '1 раздельный')] = 1
df['Ванная'].loc[(df['Ванная'] == '1 совмещенный, 1 раздельный') | (df['Ванная'] == '2 раздельных') \
    | (df['Ванная'] == '2 совмещенных') | (df['Ванная'] == '1 совмещенный, 2 раздельных')] = 2
```

```python
df
```

```python
#Для каждого найденного параметра количества ванн в квартире посчитаем среднюю площадь

ratio = [-1]*df['Ванная'].nunique()

for i in range(df['Ванная'].nunique()):
    ratio[i] = df[df['Ванная'] == i + 1]['Общая площадь'].mean()
    print(ratio[i])
```

```python
#Функция number_near будет находить ближайшее подходящее значение количества ванн для квартиры с заданной площадью
def comparison_s(n):
    pos = ratio.copy()
    pos.append(n)
    pos = sorted(pos)
    if abs(pos[pos.index(n)] - pos[pos.index(n)-1]) < abs(pos[pos.index(n)] - pos[pos.index(n)-2]):
        return ratio.index(pos[pos.index(n) - 1]) + 1
    else: 
        return ratio.index(pos[pos.index(n) - 2]) + 1
    
#Заполняем пропуски в колонке "Ванная"

s_flats = df[df['Ванная'].isna()].index
for s in s_flats:
    df['Ванная'][s] = comparison_s(df[df['Ванная'].isna()]['Общая площадь'][s])
```
